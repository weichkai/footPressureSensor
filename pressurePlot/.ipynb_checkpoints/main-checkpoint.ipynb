```python
# https://stackoverflow.com/questions/34887917/creating-a-pressure-heatmap-of-a-foot
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from matplotlib.path import Path

# Create a heat map with some fake data
ax = plt.subplot(111)
plt.pcolor(np.random.random((10,10)))

# Define a path that is a foot shape
foot_verts = [(3.2, 7.5),                         #Start point
             (2.9, 5.6), (4.1, 4.4), (4, 3.5),    #Cubic Bezier controls and end point
             (4.1,2.5), (3.9, 1.3), (4.4, 0.6),
             (5, 0), (6.3, 0.2), (6.8, 0.5),
             (7.3, 1.7), (6.7, 2.6), (6.5, 3.8),
             (6.4, 5.1), (7.6, 5.7), (7.2, 8.3),
             (6.7, 10.8), (3.5, 9.4), (3.2, 7.5)]

foot_codes = [Path.MOVETO,
              Path.CURVE4, Path.CURVE4, Path.CURVE4,
              Path.CURVE4, Path.CURVE4, Path.CURVE4,
              Path.CURVE4, Path.CURVE4, Path.CURVE4,
              Path.CURVE4, Path.CURVE4, Path.CURVE4,
              Path.CURVE4, Path.CURVE4, Path.CURVE4,
              Path.CURVE4, Path.CURVE4, Path.CURVE4]
# Define a path that is the same size as the plot area
xlim = ax.get_xlim()
ylim = ax.get_ylim()

ax_verts = [(xlim[0],ylim[0]),
            (xlim[0],ylim[1]),
            (xlim[1],ylim[1]),
            (xlim[1],ylim[0]),
            (xlim[0],ylim[0])]

ax_codes = [Path.MOVETO,
            Path.LINETO,
            Path.LINETO,
            Path.LINETO,
            Path.LINETO
            ]

#Create a patch that is the plot area minus the foot shape and add to the plot
path = Path(ax_verts + foot_verts, ax_codes + foot_codes)
patch = patches.PathPatch(path, facecolor='white', edgecolor='none')
ax.add_patch(patch)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# 假设这是你采集到的传感器数据，一个8x5的矩阵
sensor_data = np.random.rand(8, 5)  # 这里用随机数据代替

# 加载鞋垫的形状图像
insole_image = Image.open('insole_shape.png')  # 替换为你的鞋垫形状图片路径
insole_image = insole_image.convert('RGB')

# 获取鞋垫形状图像的大小
width, height = insole_image.size

# 创建一个与鞋垫形状大小相同的画布
fig, ax = plt.subplots(figsize=(width/100, height/100))

# 显示鞋垫形状图像
ax.imshow(insole_image)

# 定义每个传感器数据在鞋垫形状上的位置和大小
sensor_size = 20  # 传感器数据点的大小

# 将传感器数据绘制在鞋垫形状图像上
for i in range(sensor_data.shape[0]):
    for j in range(sensor_data.shape[1]):
        # 计算每个传感器数据点的位置
        x = j * (width / sensor_data.shape[1])
        y = i * (height / sensor_data.shape[0])
        
        # 将数据点映射到0-255的灰度值
        intensity = int(sensor_data[i, j] * 255)
        color = (intensity, intensity, intensity)
        
        # 在鞋垫形状图像上绘制传感器数据点
        circle = plt.Circle((x, y), sensor_size, color=color, alpha=0.6)
        ax.add_patch(circle)

# 设置图像标题和坐标轴
ax.set_title('Sensor Data on Insole Shape')
ax.axis('off')  # 关闭坐标轴

# 显示图像
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from matplotlib.colors import Normalize

# 假设这是你采集到的传感器数据，一个8x5的矩阵
sensor_data = np.random.rand(8, 5)  # 这里用随机数据代替

# 加载鞋垫的形状图像
insole_image = Image.open('insole_shape.png')  # 替换为你的鞋垫形状图片路径
insole_image = insole_image.convert('RGB')

# 获取鞋垫形状图像的大小
width, height = insole_image.size

# 创建一个与鞋垫形状大小相同的画布
fig, ax = plt.subplots(figsize=(width/100, height/100))

# 显示鞋垫形状图像
ax.imshow(insole_image)

# 定义每个传感器数据在鞋垫形状上的位置和大小
sensor_size = 20  # 传感器数据点的大小

# 设定颜色映射（从蓝色到红色）
norm = Normalize(vmin=0, vmax=1)
cmap = plt.cm.get_cmap('coolwarm')

# 将传感器数据绘制在鞋垫形状图像上
for i in range(sensor_data.shape[0]):
    for j in range(sensor_data.shape[1]):
        # 计算每个传感器数据点的位置
        x = j * (width / sensor_data.shape[1])
        y = i * (height / sensor_data.shape[0])
        
        # 获取传感器数据值，并映射到颜色映射范围
        value = sensor_data[i, j]
        color = cmap(norm(value))
        
        # 在鞋垫形状图像上绘制传感器数据点
        circle = plt.Circle((x, y), sensor_size, color=color, alpha=0.6)
        ax.add_patch(circle)

# 设置图像标题和坐标轴
ax.set_title('Sensor Data on Insole Shape')
ax.axis('off')  # 关闭坐标轴

# 显示图像
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from matplotlib.colors import Normalize

# 假设这是你采集到的传感器数据，一个5个传感器
sensor_data = np.random.rand(5)  # 这里用随机数据代替

# 指定每个传感器在鞋垫形状上的位置，注意要与鞋垫形状图像上的像素坐标对应
sensor_positions = [
    (100, 200),  # 传感器1的位置 (x, y)
    (150, 300),  # 传感器2的位置 (x, y)
    (200, 400),  # 传感器3的位置 (x, y)
    (250, 500),  # 传感器4的位置 (x, y)
    (300, 600)   # 传感器5的位置 (x, y)
]

# 加载鞋垫的形状图像
insole_image = Image.open('insole_shape.png')  # 替换为你的鞋垫形状图片路径
insole_image = insole_image.convert('RGB')

# 获取鞋垫形状图像的大小
width, height = insole_image.size

# 创建一个与鞋垫形状大小相同的画布
fig, ax = plt.subplots(figsize=(width/100, height/100))

# 显示鞋垫形状图像
ax.imshow(insole_image)

# 设定颜色映射（从蓝色到红色）
norm = Normalize(vmin=0, vmax=1)
cmap = plt.cm.get_cmap('coolwarm')

# 将传感器数据绘制在鞋垫形状图像上
for i, (x, y) in enumerate(sensor_positions):
    # 获取传感器数据值，并映射到颜色映射范围
    value = sensor_data[i]
    color = cmap(norm(value))
    
    # 在鞋垫形状图像上绘制传感器数据点
    circle = plt.Circle((x, y), 10, color=color, alpha=0.6)
    ax.add_patch(circle)

# 设置图像标题和坐标轴
ax.set_title('Sensor Data on Insole Shape')
ax.axis('off')  # 关闭坐标轴

# 显示图像
plt.tight_layout()
plt.show()
```

```python
# https://stackoverflow.com/questions/34887917/creating-a-pressure-heatmap-of-a-foot
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from matplotlib.path import Path

# Create a heat map with some fake data
ax = plt.subplot(111)
plt.pcolor(np.random.random((10,10)))

# Define a path that is a foot shape
foot_verts = [(3.2, 7.5),                         #Start point
             (2.9, 5.6), (4.1, 4.4), (4, 3.5),    #Cubic Bezier controls and end point
             (4.1,2.5), (3.9, 1.3), (4.4, 0.6),
             (5, 0), (6.3, 0.2), (6.8, 0.5),
             (7.3, 1.7), (6.7, 2.6), (6.5, 3.8),
             (6.4, 5.1), (7.6, 5.7), (7.2, 8.3),
             (6.7, 10.8), (3.5, 9.4), (3.2, 7.5)]

foot_codes = [Path.MOVETO,
              Path.CURVE4, Path.CURVE4, Path.CURVE4,
              Path.CURVE4, Path.CURVE4, Path.CURVE4,
              Path.CURVE4, Path.CURVE4, Path.CURVE4,
              Path.CURVE4, Path.CURVE4, Path.CURVE4,
              Path.CURVE4, Path.CURVE4, Path.CURVE4,
              Path.CURVE4, Path.CURVE4, Path.CURVE4]
# Define a path that is the same size as the plot area
xlim = ax.get_xlim()
ylim = ax.get_ylim()

ax_verts = [(xlim[0],ylim[0]),
            (xlim[0],ylim[1]),
            (xlim[1],ylim[1]),
            (xlim[1],ylim[0]),
            (xlim[0],ylim[0])]

ax_codes = [Path.MOVETO,
            Path.LINETO,
            Path.LINETO,
            Path.LINETO,
            Path.LINETO
            ]

#Create a patch that is the plot area minus the foot shape and add to the plot
path = Path(ax_verts + foot_verts, ax_codes + foot_codes)
patch = patches.PathPatch(path, facecolor='white', edgecolor='none')
ax.add_patch(patch)

plt.show()
```
